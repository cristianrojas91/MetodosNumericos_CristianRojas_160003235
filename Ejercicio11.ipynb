```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import math
import sys
import matplotlib.pyplot as pl

x1=[]
y1=[]
x2=[]
y2=[]

def funcion():
    
    global x1,y1,x2,y2
    x=3.0
    while(x<=5.0):
        
        total1 = ((-3.0/2.0)*x) + 9.0
        total2 = (x/2.0) + 1.0
        x1.append(x)
        y1.append(total1)
        x2.append(x)
        y2.append(total2)
        x=x+0.5
        
def graficar():
    
    global x1,y1,x2,y2
    pl.plot(x1,y1,"r", label="funcion1")
    pl.plot(x2,y2,"b", label="funcion2")
    pl.title("Solucion Grafica")
    pl.legend(loc="upper center")
    pl.grid()
    pl.show()
            
def main():
    funcion()
    graficar()
    
if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import math
import sys
import matplotlib.pyplot as pl

x1=[]
y1=[]
y2=[]
x2=[]
y3=[]
y4=[]
x3=[]
y5=[]
y6=[]

def funcion1():
    
    global x1,y1,y2
    x=-2.0
    while(x<=7.0):
        
        total1 = 1.0 + (1.0/2.0)*x
        total2 = (1.0/2.0) + (1.0/2.0)*x
        x1.append(x)
        y1.append(total1)
        y2.append(total2)
        x=x+0.5
        
def funcion2():
    
    global x2,y3,y4
    x=-2.0
    while(x<=7.0):
        
        total3 = 1.0 + (1.0/2.0)*x
        total4 = 1.0 + (1.0/2.0)*x
        x2.append(x)
        y3.append(total3)
        y4.append(total4)
        x=x+0.5
        
def funcion3():
    
    global x3,y5,y6
    x=-2.0
    while(x<=7.0):
        
        total5 = 1.1 + (2.3/5.0)*x
        total6 = 1.0 + (1.0/2.0)*x
        x3.append(x)
        y5.append(total5)
        y6.append(total6)
        x=x+0.5
        
def graficar():
    
    global x1,y1,y2,x2,y3,y4,x3,y5,y6
    pl.figure(figsize=(16,8))
    
    pl.subplot(2,2,1)
    pl.plot(x1,y1,"r", label="funcion1")
    pl.plot(x1,y2,"b", label="funcion2")
    pl.title("a) No hay solucion")
    pl.legend(loc="upper center")
    pl.grid()
    
    pl.subplot(2,2,2)
    pl.plot(x2,y3,"r", label="funcion3")
    pl.plot(x2,y4,"b", label="funcion4")
    pl.title("b) Infinitas soluciones")
    pl.legend(loc="upper center")
    pl.grid()
    
    pl.subplot(2,2,3)
    pl.plot(x3,y5,"r", label="funcion5")
    pl.plot(x3,y6,"b", label="funcion6")
    pl.title("c) Sistema mal condicionado")
    pl.legend(loc="upper center")
    pl.grid()
    
    pl.show()
            
def main():
    funcion1()
    funcion2()
    funcion3()
    graficar()
    
if __name__ == "__main__":
    main()
```
