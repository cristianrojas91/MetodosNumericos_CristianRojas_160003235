```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import math
import sys
import matplotlib.pyplot as pl

x1=[]
y1=[]
x2=[]
y2=[]
x3=[]
y3=[]
y4=[]
y5=[]
y6=[]
x4=[]
y7=[]
y8=[]

def raphson():
    
    global x1,y1,y4
    print "Newton-Raphson\n"
    print "i\t Total\t\t Ea\t\t Et\n"
    Es = 0.00000005
    i = 0
    xi = 0
    anterior = 0.0
    Ea = 1.0
    verdadero= 0.56714329
    
    while Ea>=Es:
        
        aprox = xi - ((math.exp(-xi)-xi)/(-math.exp(-xi)-1))
        Ea = abs(((aprox-anterior)/aprox)*100)
        Et = abs(((verdadero-xi)/verdadero)*100)
        print i, "\t%.10f"%xi, "\t%.10f"%Ea, "\t%.10f"%Et
        anterior = aprox
        xi = aprox
        x1.append(i)
        y1.append(Ea)
        y4.append(Et)
        i=i+1
        
        
def secante():
    
    global x2,y2,y5
    print "\nSecante"
    print "\ni\t Total\t\t Ea\t\t Et\n"
    Es = 0.00000005
    i = 0
    xi = 1
    anterior = 0.0
    Ea = 1.0
    verdadero= 0.56714329
    
    while Ea>=Es:
        
        aprox = xi - ((math.exp(-xi)-xi)*(anterior-xi))/((math.exp(-anterior)-anterior)-(math.exp(-xi)-xi))
        Ea = abs(((aprox-anterior)/aprox)*100)
        Et = abs(((verdadero-xi)/verdadero)*100)
        print i, "\t%.10f"%aprox, "\t%.10f"%Ea, "\t%.10f"%Et
        anterior = xi
        xi = aprox
        x2.append(i)
        y2.append(Ea)
        y5.append(Et)
        i=i+1
        
def puntoFijo():
    
    global x3,y3,y6
    Es=0.0005
    Ea=1.0
    i=0
    x=0.0
    real = 0.56714329
    print "\nPunto Fijo"
    print "\ni\t Total\t\t Ea\t\t Et\n"
    while(Ea>=Es):
        
        total = math.exp(-x) 
        
        Ea = abs(((total-x)/total)*100)
        Et = abs(((real-total)/real)*100)
        
        print i, "\t%.10f"%total, "\t%.10f"%Ea, "\t%.10f"%Et
        x = total
        x3.append(i)
        y3.append(Ea)
        y6.append(Et)
        i=i+1

def f(x):
    
    valor = (math.exp(-x))-x
    return valor

def biseccion():
    
    xl=0.2
    xu=0.8
    Es=1*(10**(-8))
    Ea= 1.0
    anterior=0
    real=0.56714329
    i=0
    
    print "\nBiseccion"
    print "\ni\t Total\t\t Ea\t\t Et\n"
    
    if(f(xl)*f(xu)<0):
        
        while(Ea>=Es):
        
            xr = (xl+xu)/2

            if(f(xl)*f(xr)<0):
                xu=xr
                Ea = abs(((xr-anterior)/xr)*100)
                Et = abs(((real-xu)/real)*100)
                anterior = xr
                print i, "\t%.10f"%xu, "\t%.10f"%Ea, "\t%.10f"%Et
                x4.append(i)
                y7.append(Ea)
                y8.append(Et)
                i=i+1
                
            if(f(xl)*f(xr)>0):
                
                xl=xr
                Ea = abs(((xr-anterior)/xr)*100)
                Et = abs(((real-xl)/real)*100)
                anterior = xr
                print i, "\t%.10f"%xl, "\t%.10f"%Ea, "\t%.10f"%Et
                x4.append(i)
                y7.append(Ea)
                y8.append(Et)
                i=i+1
                
            if(f(xl)*f(xr)==0):
                print "raiz\t", xr
                break


def graficar():
    
    global x1,y1,x2,y2,x3,y3,y4,y5,y6
    pl.figure(figsize=(16,8))
    
    pl.subplot(211)
    pl.plot(x1,y1,"r", label="Raphson")
    pl.plot(x2,y2,"g", label="Secante")
    pl.plot(x3,y3,"b", label="PuntoFijo")
    pl.plot(x4,y7,"m", label="Biseccion")
    pl.title("Ea")
    pl.legend(loc="upper center")
    pl.grid()
    
    pl.subplot(212)
    pl.plot(x1,y4,"r", label="Raphson")
    pl.plot(x2,y5,"g", label="Secante")
    pl.plot(x3,y6,"b", label="PuntoFijo")
    pl.plot(x4,y8,"m", label="Biseccion")
    pl.title("Et")
    pl.legend(loc="upper center")
    pl.grid()
    
    pl.show() 

    
def main():
    raphson()
    secante()
    puntoFijo()
    biseccion()

if __name__ == "__main__":
    main()
    graficar()
```
