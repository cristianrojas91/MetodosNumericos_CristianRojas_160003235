```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import math
import sys
import matplotlib.pyplot as pl

x1=[]
y1=[]
y2=[]
y3=[]
y4=[]
y5=[]
y6=[]
y7=[]
y8=[]
y9=[]
y10=[]
y11=[]
y12=[]
y13=[]
y14=[]
y15=[]

def funcion1(x):
    total = ((math.e)**x)-1
    return total

def funcion2(x):
    total = (math.sin(2*x)**3)+(x**2)
    return total

def funcion3(x):
    total = (x**4)+(3*(x**2))-math.cos(x)
    return total

def calculo(h):
    
    x = -5
    while(x<=5):
        
        
        derivada1 = (math.e)**x
        derivada2 = 6*((math.sin(2*x))**2)*(math.cos(2*x))+(2*x)
        derivada3 = 4*(x**3)+(6*x)+math.sin(x)
        
        f1 = (funcion1(x+h)-funcion1(x))/h
        f2 = (funcion1(x)-funcion1(x-h))/h
        f3 = (funcion1(x+h)-funcion1(x-h))/(2*h)

        f4 = (funcion2(x+h)-funcion2(x))/h
        f5 = (funcion2(x)-funcion2(x-h))/h
        f6 = (funcion2(x+h)-funcion2(x-h))/(2*h)

        f7 = (funcion3(x+h)-funcion3(x))/h
        f8 = (funcion3(x)-funcion3(x-h))/h
        f9 = (funcion3(x+h)-funcion3(x-h))/(2*h)

        ev1 = derivada1-f1
        ev2 = derivada1-f2
        ev3 = derivada1-f3

        ev4 = derivada2-f4
        ev5 = derivada2-f5
        ev6 = derivada2-f6

        ev7 = derivada3-f7
        ev8 = derivada3-f8
        ev9 = derivada3-f9
            
        try:
            if(derivada1==0):
                derivada1=0.0000000001
            if(derivada2==0):
                derivada2=0.0000000001
            if(derivada3==0):
                derivada3=0.0000000001

            
            erp1 = ((derivada1-f1)/derivada1)*100
            erp2 = ((derivada1-f2)/derivada1)*100
            erp3 = ((derivada1-f3)/derivada1)*100

            erp4 = ((derivada2-f4)/derivada2)*100
            erp5 = ((derivada2-f5)/derivada2)*100
            erp6 = ((derivada2-f6)/derivada2)*100

            erp7 = ((derivada3-f7)/derivada3)*100
            erp8 = ((derivada3-f8)/derivada3)*100
            erp9 = ((derivada3-f9)/derivada3)*100
            
        except:
            print "\nError! Division por Cero!"
            
        print "\n---------------------------------------------------------------------------------------------------------"
        print "  X =","%.2f"%x
        print "---------------------------------------------------------------------------------------------------------"
        print "Funcion: (e^x)-1"
        print "Derivada Real: ", "%.4f" % derivada1
        print "Atras: ", "\t%.4f" % f2," *ErrorVer:","%.4f"%ev2," *ErrorVerAbs:","%.4f"%abs(ev2)," *ErrorRelPor:","%.4f"%abs(erp2),"%"
        print "Centrada: ", "\t%.4f" % f3," *ErrorVer:","%.4f"%ev3," *ErrorVerAbs:","%.4f"%abs(ev3)," *ErrorRelPor:","%.4f"%abs(erp3),"%"
        print "Adelante: ", "\t%.4f" % f1," *ErrorVer:","%.4f"%ev1," *ErrorVerAbs:","%.4f"%abs(ev1)," *ErrorRelPor:","%.4f"%abs(erp1),"%"
        print "---------------------------------------------------------------------------------------------------------"
        print "Funcion: (sen(2x)^3)+(x^2)"
        print "Derivada Real: ", "%.4f" % derivada2
        print "Atras: ", "\t%.4f" % f5," *ErrorVer:","%.4f"%ev5," *ErrorVerAbs:","%.4f"%abs(ev5)," *ErrorRelPorc:","%.4f"%abs(erp5),"%"
        print "Centrada: ", "\t%.4f" % f6," *ErrorVer:","%.4f"%ev6," *ErrorVerAbs:","%.4f"%abs(ev6)," *ErrorRelPorc:","%.4f"%abs(erp6),"%"
        print "Adelante: ", "\t%.4f" % f4," *ErrorVer:","%.4f"%ev4," *ErrorVerAbs:","%.4f"%abs(ev4)," *ErrorRelPorc:","%.4f"%abs(erp4),"%"
        print "---------------------------------------------------------------------------------------------------------"
        print "Funcion: (x^4)+(3(x^2))-cos(x)"
        print "Derivada Real: ", "%.4f" % derivada3
        print "Atras: ", "\t%.4f" % f8," *ErrorVer:","%.4f"%ev8," *ErrorVerAbs:","%.4f"%abs(ev8)," *ErrorRelPor:","%.4f"%abs(erp8),"%"
        print "Centrada: ", "\t%.4f" % f9," *ErrorVer:","%.4f"%ev9," *ErrorVerAbs:","%.4f"%abs(ev9)," *ErrorRelPor:","%.4f"%abs(erp9),"%"
        print "Adelante: ", "\t%.4f" % f7," *ErrorVer:","%.4f"%ev7," *ErrorVerAbs:","%.4f"%abs(ev7)," *ErrorRelPor:","%.4f"%abs(erp7),"%"
        
        x1.append(x)
        
        y1.append(f1)
        y2.append(f2)
        y3.append(f3)
        y4.append(derivada1)
        y5.append(erp3)
        
        y6.append(f4)
        y7.append(f5)
        y8.append(f6)
        y9.append(derivada2)
        y10.append(erp6)
        
        y11.append(f7)
        y12.append(f8)
        y13.append(f9)
        y14.append(derivada3)
        y15.append(erp9)
        x=x+h;
        
def graficar():
    
    global x1,y1,y2,y3,y4,y5,y6,y7,y8,y9,y10,y11,y12,y13,y14,y15
    pl.figure(figsize=(16,8))
    
    pl.subplot(2,2,1)
    pl.plot(x1,y1,"y", label="Adelante")
    pl.plot(x1,y2,"g", label="Atras")
    pl.plot(x1,y3,"b", label="Centrada")
    pl.plot(x1,y4,"r", label="Real")
    pl.title("Funcion1: (e^x)-1")
    pl.legend(loc="upper left")
    pl.grid()
    
    pl.subplot(2,2,2)
    pl.plot(x1,y6,"y", label="Adelante")
    pl.plot(x1,y7,"g", label="Atras")
    pl.plot(x1,y8,"b", label="Centrada")
    pl.plot(x1,y9,"r", label="Real")
    pl.title("Funcion2: (sen(2x)^3)+(x^2)")
    pl.legend(loc="upper left")
    pl.grid()
    
    pl.subplot(2,2,3)
    pl.plot(x1,y11,"y", label="Adelante")
    pl.plot(x1,y12,"g", label="Atras")
    pl.plot(x1,y13,"b", label="Centrada")
    pl.plot(x1,y14,"r", label="Real")
    pl.title("Funcion3: (x^4)+(3(x^2))-cos(x)")
    pl.legend(loc="upper left")
    pl.grid()
    
    pl.subplot(2,2,4)
    pl.plot(x1,y5,"r", label="Funcion1")
    pl.plot(x1,y10,"b", label="Funcion2")
    pl.plot(x1,y15,"g", label="Funcion3")
    pl.title("Error Relativo Procentual")
    pl.legend(loc="upper left")
    pl.grid()
    
    pl.show()
    
def main():
    
    h = input("Ingrese el valor de h: ")
    calculo(h)
    

if __name__ == "__main__":
    main()
    graficar()
```
