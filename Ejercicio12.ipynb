# Ejercicio 9.4

Metodo grafico:

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import math
import sys
import matplotlib.pyplot as pl

x1=[]
y1=[]
x2=[]
y2=[]

def funcion():
    
    global x1,y1,x2,y2
    x=3.0
    while(x<=5.0):
        
        total1 = ((1.0/2.0)*x) + 3.0
        total2 = ((-1.0/6.0)*x) + (34.0/6.0)
        x1.append(x)
        y1.append(total1)
        x2.append(x)
        y2.append(total2)
        x=x+0.5
        
def graficar():
    
    global x1,y1,x2,y2
    pl.plot(x1,y1,"r", label="funcion1")
    pl.plot(x2,y2,"b", label="funcion2")
    pl.title("Solucion Grafica")
    pl.legend(loc="upper center")
    pl.grid()
    pl.show()
            
def main():
    funcion()
    graficar()
    
if __name__ == "__main__":
    main()
```
